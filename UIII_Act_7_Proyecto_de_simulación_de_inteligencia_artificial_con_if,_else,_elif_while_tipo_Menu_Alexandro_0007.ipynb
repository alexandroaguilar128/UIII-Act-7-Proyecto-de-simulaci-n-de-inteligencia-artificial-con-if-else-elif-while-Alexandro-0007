### UIII Act 7 Proyecto de simulación de inteligencia artificial con if, else, elif  while tipo Menu.
### Medico y Paciente

```python
atendiendo = True
nombre_paciente = "No registrado"
sintoma = "Ninguno"
temperatura = 36.5

print("--- BIENVENIDO AL CONSULTORIO VIRTUAL IA ---")

while atendiendo == True:
    print("\nMENÚ DE ATENCIÓN:")
    print("1. Registrar nombre del paciente")
    print("2. Ingresar síntomas y signos")
    print("3. Obtener diagnóstico de la IA")
    print("4. Salir de la consulta")

    opcion = input("Seleccione una opción (1-4): ")

    # CASO 1: Registro del paciente
    if opcion == "1":
        nombre_paciente = input("IA: ¿Cuál es el nombre del paciente?: ")
        print("IA: Paciente " + nombre_paciente + " registrado correctamente.")

    # CASO 2: Captura de síntomas
    elif opcion == "2":
        sintoma = input("IA: Describa el síntoma principal (ej. tos, dolor, cansancio): ")
        temperatura = float(input("IA: Ingrese la temperatura corporal en grados Celsius: "))
        print("IA: Datos guardados para el análisis.")

    # CASO 3: Simulación de Inteligencia Artificial (Diagnóstico) [18.8.6]
    elif opcion == "3":
        print("IA: Analizando datos de " + nombre_paciente + "...")

        # Lógica de decisión con operadores lógicos y relacionales [5, 6]
        if temperatura > 38.0 and sintoma == "tos":
            print("IA DIAGNÓSTICO: Es muy probable que tenga una infección respiratoria o gripe.")
            print("IA RECOMIENDA: Reposo absoluto y mucha hidratación.")

        elif temperatura > 37.5 and sintoma == "dolor":
            print("IA DIAGNÓSTICO: Presenta febrícula y dolor localizado. Podría ser una inflamación.")
            print("IA RECOMIENDA: Monitorear la temperatura cada 4 horas.")

        elif temperatura <= 37.0 and sintoma == "Ninguno":
            print("IA DIAGNÓSTICO: El paciente parece estar en perfecto estado de salud.")

        else:
            print("IA DIAGNÓSTICO: Síntomas no concluyentes para mi base de datos actual.")
            print("IA RECOMIENDA: Consulte a un médico humano para una revisión física.")

    # CASO 4: Salida del programa [7]
    elif opcion == "4":
        print("IA: Finalizando sesión médica. ¡Cuide su salud, " + nombre_paciente + "!")
        atendiendo = False
print("Programa elaborado por Alexandro Aguilar NL = 1")
```
